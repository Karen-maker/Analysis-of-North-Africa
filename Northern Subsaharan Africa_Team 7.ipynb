Northern Sub-Saharan Africa   


Northern Sub-Saharan Africa belongs to the northern part of the African Continent which is mostly represented by a vast area over the Sahara Desert. Its economy is mostly based on:  

·Mineral Deposits - petroleum, phosphate, and natural gases  
·Agriculture – cotton, Arabic gum, cork, olives, figs  
·Camels  
·Tourism 

Based on the natural, available resources, indigenous have preferred focusing on developing and exploiting primary industries and processing industries for petroleum and phosphate rock. It is part of the developing countries and shows a low but increasing economic growth rate. The unemployment rate is very high, and stats demonstrate that there are not enough jobs to effectively reduce it to improve people’s livelihood.   
Most of the population is white and the main religion is Islamism, followed by Christianity. Northern Sub-Saharan Africa is famous for its long coastline and scenic spots like the Sahara Desert, the World Cultural Heritage, Historic Mosques, Roman Ruins and so on.

The following step consisted of choosing out of the total 41 columns, one or two per category based on missing values and importance (For instance instead of choosing the Employment ratio of Male or Female we took into account the Total Employment to population ratio to have a better overview, data, and an overall analysis gender-free). The final skimming let us end up with the top 9 columns representing the categories that perfectly suit our quantitative and qualitative analysis.  
For what concerns the beginning of our analysis, with the help of Python’s function “isnull().sum()” we were able to identify the missing values turning out to be 72, spread among 17 columns. Hence, we investigated the given data to see if we can trust it, and with it fill in the missing values using the median, the mean, or with data from external research.      
Being our rows limited to 10, columns with more than 30% of missing values were not considered. Out of our 9 selected columns we only had one presenting missing values, “Primary completion rate, total (% of relevant age group)” under the Education segment.     
We decided to substitute the missing values with either mean or median. To choose the best suitable substitute we used an imputation method by creating a histogram that showed a right skewness of our median.

Considering the metadata year, we found out that each Country in our Region has different years used to gather the data, which means, the mean value was collected at different time frames. So, we did external research to compare the given data in that specific year with the data from the research.  
Taking Algeria as an example, Algeria’s data was collected in 2008, hence we did research based on the 9 selected columns to see if the data was reliable. According to the "[World Bank Data](https://data.worldbank.org/indicator/NE.TRD.GNFS.ZS?locations=DZ)", we can trust the numbers given to Algeria. 
We went through this process with every Country in our Region and the only one that we had some discrepancies on the data was Morocco.   
Hence using the median turned out to be the best option, not considering different times affecting our results.  
With the help of Python, we filled the NAN values exploiting “fillna” function to then select which Country would best represent our Region.  
Comparing the mean of every selected column to the values of every Country, we were able to identify Morocco as the best representative of the Northern Sub-Saharan Africa Region, on average.    
A very small difference - about 2% - was identified when comparing Morocco’s values to the mean of every column selected. For instance, based on Tuberculosis death rate, the column mean was found to be 15.2 and Morocco’s Tuberculosis death rate is 12.00. The 3.2 difference is the smallest difference when compared to the other countries.    
After seeing this difference with the means in Morocco’s values, additional research was implemented to check whether the given data and the real data were the same with the aim of investigating if the selected Country is not based on false data.
