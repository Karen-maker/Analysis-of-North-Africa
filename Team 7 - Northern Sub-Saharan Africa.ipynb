```python
# Imputting missing values

# Step 1: Drop missing values from the Dataset so that they don't cause undue complications when generation plots

# Making new Dataframe independent
df_dropped = pd.DataFrame.copy(cleaned_data_team7)

# Dropping missing values using dropna() for df_dropped
df_dropped = df_dropped.dropna().round(decimals = 2)

# Checking to see if all missing values have been dropped
print(df_dropped)
```

```python
# Visualizing Data in Python - BASED ON COLUMN #1

# setting figure size
fig, ax = plt.subplots(figsize = [8, 5])


# histogram for carat
sns.distplot(a     = (df_dropped ['Tuberculosis']),
             bins  = 'fd',   #to be changed if histogram shows gaps
             hist  = True,
             kde   = False,
             rug   = False,
             color = 'black')


# titles and labels
plt.title(label = "Distribution of AIDS deaths")
plt.xlabel(xlabel = 'AIDS deaths')
plt.ylabel(ylabel = 'Frequency')


# New: These add vertical lines to the code
plt.axvline(x = (df_dropped ['Tuberculosis']).mean(),
            color = 'red')


plt.axvline(x = (df_dropped ['Tuberculosis']).median(),
            color = 'blue')


# this adds a legend
plt.legend(labels =  ['mean', 'median'])


# these compile and display the plot so that it is formatted as expected
plt.tight_layout()
plt.show()
```
